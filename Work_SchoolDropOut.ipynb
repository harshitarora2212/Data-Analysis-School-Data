```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Calling the csv
main=pd.read_csv('data-ratio-2012-2015.csv')
```

```python
# Making a Backup Copy
data=main.copy()
```

```python
data.shape
```

```python
data.head()
```

```python
data.columns.value_counts()
```

```python
data.isna().sum()
# No Null Values
```

```python
data['Primary_Boys'].values
```

```python
# Replacing the Invalid values
data=data.replace('NR',0)
```

```python
data['Primary_Boys'].values
```

```python
# Converting the values to the float from int
data['Primary_Boys']=data['Primary_Boys'].apply(lambda x:float(x))

# Alternative data['Primary_Boys].astype(float)
```

```python
data.info()
```

```python
columns=['Primary_Girls','Upper Primary_Boys','Upper Primary_Girls','Secondary _Boys','Secondary _Girls','HrSecondary_Boys','HrSecondary_Girls']

data[columns]=data[columns].apply(pd.to_numeric,errors='coerce')
# or data[columns]=pd.to_numeric(data[columns],errors='coerce')
```

```python
data.info()
```

```python
data['State_UT'].value_counts()
```

```python
plt.figure(figsize= (20,15))
plt.subplot(4,4,1)
sns.boxplot(data['HrSecondary_Girls'])

plt.subplot(4,4,2)
sns.boxplot(data['HrSecondary_Boys'])

plt.subplot(4,4,3)
sns.boxplot(data['Upper Primary_Boys'])

plt.subplot(4,4,4)
sns.boxplot(data['Upper Primary_Girls'])

plt.subplot(4,4,5)
sns.boxplot(data['Secondary _Girls'])

plt.subplot(4,4,6)
sns.boxplot(data['Secondary _Boys'])
```

```python
data['HrSecondary_Girls']=data['HrSecondary_Girls'].clip(lower=data['HrSecondary_Girls'].quantile(0.05), upper=data['HrSecondary_Girls'].quantile(0.95))

data['HrSecondary_Boys']=data['HrSecondary_Boys'].clip(lower=data['HrSecondary_Boys'].quantile(0.05), upper=data['HrSecondary_Boys'].quantile(0.95))

data['Upper Primary_Boys']=data['Upper Primary_Boys'].clip(lower=data['Upper Primary_Boys'].quantile(0.05), upper=data['Upper Primary_Boys'].quantile(0.95))

data['Upper Primary_Girls']=data['Upper Primary_Girls'].clip(lower=data['Upper Primary_Girls'].quantile(0.05), upper=data['Upper Primary_Girls'].quantile(0.95))

data['Secondary _Girls']=data['Secondary _Girls'].clip(lower=data['Secondary _Girls'].quantile(0.05), upper=data['Secondary _Girls'].quantile(0.95))

data['Secondary _Boys']=data['Secondary _Boys'].clip(lower=data['Secondary _Boys'].quantile(0.05), upper=data['Secondary _Boys'].quantile(0.95))
```

```python
plt.figure(figsize= (20,15))
plt.subplot(4,4,1)
sns.boxplot(data['HrSecondary_Girls'])

plt.subplot(4,4,2)
sns.boxplot(data['HrSecondary_Boys'])

plt.subplot(4,4,3)
sns.boxplot(data['Upper Primary_Boys'])

plt.subplot(4,4,4)
sns.boxplot(data['Upper Primary_Girls'])

plt.subplot(4,4,5)
sns.boxplot(data['Secondary _Girls'])

plt.subplot(4,4,6)
sns.boxplot(data['Secondary _Boys'])
```
